# Crop Area-Estimation

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
import cv2
from collections import Counter
from skimage.color import rgb2lab, deltaE_cie76
import os

%matplotlib inline
```

```python
# function for RGB to HEX
def RGB2HEX(color):
    return "#{:02x}{:02x}{:02x}".format(int(color[0]), int(color[1]), int(color[2]))
```

```python
# Function to read image
# cv2 reads images in BGR format, hence conversion back to RGB
# is done here.
def get_image(image_path):
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    return image
```

```python
# Function to retrive colours from image

def get_area_per(image, number_of_colors, show_chart):
    
    modified_image = cv2.resize(image, (600, 400), interpolation = cv2.INTER_AREA)
    # Flatten image so to process via KMEANS
    modified_image = modified_image.reshape(modified_image.shape[0]*modified_image.shape[1], 3)
    
    clf = KMeans(n_clusters = number_of_colors)
    labels = clf.fit_predict(modified_image)
    
    counts = Counter(labels)
    # sort to ensure correct color percentage
    counts = dict(sorted(counts.items()))
    
    center_colors = clf.cluster_centers_
    # We get ordered colors by iterating through the keys
    ordered_colors = [center_colors[i] for i in counts.keys()]
    hex_colors = [RGB2HEX(ordered_colors[i]) for i in counts.keys()]
    rgb_colors = [ordered_colors[i] for i in counts.keys()]

    if (show_chart):
        plt.figure(figsize = (8, 6))
        plt.pie(counts.values(), labels = hex_colors, colors = hex_colors)
    
    # Area percentage Estimation
    color_ctr = list(counts.values())
    total_area = sum(color_ctr)
    area_per = []
    for i in range(len(hex_colors)):
        area_per.append({"per":(color_ctr[i]/total_area)*100, "rgb":rgb_colors[i], 'hex':hex_colors[i]})
    return area_per
```

```python
def match_color_per(image, color, threshold = 60, number_of_colors = 8): 
    
    image_colors = get_area_per(get_image(image), number_of_colors, True)
    selected_color = rgb2lab(np.uint8(np.asarray([[color]])))

    area = 0;
    
    select_image = False
    for i in range(number_of_colors):
        curr_color = rgb2lab(np.uint8(np.asarray([[image_colors[i]["rgb"]]])))
        diff = deltaE_cie76(selected_color, curr_color)
        if (diff < threshold):
            area += image_colors[i]["per"]
    
    return area
```
